```python
# ASSIGNMENT 1
# head()
import pandas as pd
df = pd.read_csv('work.csv')
# The head() method returns the headers and a specified number of rows, starting from the top.
print(df.head(4))
```

```python
#tail()
# The tail() method returns the headers and a specified number of rows, starting from the bottom.
print(df.tail(3))
```

```python
# info()
#  info(), that gives you more information about the data set.
print(df.info()) 
```

```python
# dropna()
# this function is used to return a new Data Frame with no empty cells
df1 = df.dropna()

print(df1.to_string())
```

```python
# fillna()
# The fillna() method allows us to replace empty cells with a value
# Replace NULL values with the number 100
df.fillna(100, inplace = True)
print(df)
```

```python
# removing wrong data by replacing values
df.loc[3, 'city'] = "Ibanda"
df.loc[5, 'age'] = 22
df.loc[2, 'py-score'] = 99

print(df)

```

```python
# loading mine.csv dataset
import pandas as pd
df2 = pd.read_csv('mine.csv')
print(df2)
```

```python
# wrong format for large dataset
# To replace wrong data for larger data sets you can create some boundaries for legal values,
#and replace any values that are outside of the boundaries.
for x in df2.index:
  if df2.loc[x, "Duration"] > 100:
    df2.loc[x, "Duration"] = 100
print(df2.to_string())
```

```python
# Another way of handling wrong data is to remove the rows that contains wrong data
for x in df2.index:
  if df2.loc[x, "Duration"] > 99:
    df2.drop(x, inplace = True)
print(df2.to_string())
#remember to include the 'inplace = True' argument to make the changes in the original
#DataFrame object instead of returning a copy

```

```python
# removing wrong data (NaN)
df2.dropna(subset=['Calories'], inplace = True)
print(df2.to_string())
```

```python
# discovering duplicates in our dataset
#To discover duplicates, we can use the duplicated() method which returns a Boolean values for each row
print(df2.duplicated())
```

```python
# To remove duplicates, use the drop_duplicates() method
df2.drop_duplicates(inplace = True)

print(df2.to_string())
```

```python
# performing correlations
# using corr()
#The corr() method calculates the relationship between each column in your data set.
#Perfect Correlation =1.000000
#Good Correlation =0.806032
#bad Correlation =0.085104
print(df2.corr())
```
