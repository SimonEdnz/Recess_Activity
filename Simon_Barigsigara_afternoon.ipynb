```python
# qtn5
import pandas as pd

# Read the Excel file and specify the column name and the engine
data_frame = pd.read_csv('AfricaCupofNationsMatches.csv')
column_name = 'HomeTeamGoals'

# Convert the column to a Pandas series
series = data_frame[column_name]

# Calculate the frequency of each value
value_counts = series.value_counts()

# Find the most frequent value
most_frequent_value = value_counts.idxmax()

# Replace all other values with 'Other'
series = series.apply(lambda x: x if x == most_frequent_value else 'Other')


print("Most frequent value:", most_frequent_value)
print("Modified series:")
print(series)

```
